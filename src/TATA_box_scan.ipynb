```python
import pandas as pd
from pybedtools import BedTool
```

```python
motifs_mapped = '../data/output/non-overlapping_includingbidirectional_all_genes_newannotation/FIMO/promoters_5UTR_motifs_plantpan.bed'
plantpan_ID_mapping = '../data/FIMO/motif_data/ID_mapping_Arabidopsis_thaliana.txt'
promoters_bed = '../data/output/non-overlapping_includingbidirectional_all_genes_newannotation/FIMO/promoters_5UTR.bed'
renamed_TATA_bed = '../data/EPD_promoter_analysis/TATA/TATA_boxes_renamed.bed'
EPD_TATA_mapped = '../data/output/non-overlapping_includingbidirectional_all_genes_newannotation/TATA/EPD_TATA_mapped.bed'
```

```python
def remove_characters_linestart(input_location,output_location,oldcharacters,newcharacters,linestart):
    """this function removes characters from the start of each line in the input file and sends modified lines to output"""
    output = open(output_location, 'w') #make output file with write capability
    #open input file
    with open(input_location, 'r') as infile:  
        #iterate over lines in file
        for line in infile:
            line = line.strip() # removes hidden characters/spaces
            if line[0] == linestart:
                                 
                line = line.replace(oldcharacters, newcharacters) #remove characters from start of line, replace with new characters        
            output.write(line + '\n') #output to new file
    output.close()
    
```

```python
motifs_df = pd.read_table(motifs_mapped, sep='\t', header=None)
cols = ['chr', 'start', 'stop', 'name_rep', 'score', 'strand', 'promoter_AGI', 'p-value', 'q-value', 'matched_sequence']
motifs_df.columns = cols
```

```python
ID_mapping = pd.read_table(plantpan_ID_mapping, sep='\t', header=0)
cols2 = ['name_rep', 'TF_AGI', 'matrix_type']
ID_mapping.columns = cols2
```

```python
#merge to map ID mapping to motifs
merged = pd.merge(motifs_df, ID_mapping, on='name_rep', how='left')
```

```python
#Genes with TATA boxes:
TBP2 = merged[merged.TF_AGI == 'AT1G55520']
TBP2
```

```python
#genes with more than one TATA
TBP2_counts = TBP2.copy()
TBP2_counts['number_of_TATA'] = TBP2.groupby('promoter_AGI')['promoter_AGI'].transform('count')
```

```python
#how many TATA boxes from EPD overlap my promoters
# proms_df = pd.read_table(promoters_bed, sep='\t', header=None)
# cols1 = ['chr','start','stop','AGI','dot1','strand','source','type','dot2','attributes']
# proms_df.columns = cols1
proms = BedTool(promoters_bed)
TATA = BedTool(renamed_TATA_bed)
intersect = proms.intersect(TATA, wo=True)
#Write to output_file
with open(EPD_TATA_mapped, 'w') as output:
    #Each line in the file contains gff entry a and gff entry b that it overlaps plus the number of bp in the overlap so 19 columns
    output.write(str(intersect))    
```

```python
EPD_TATA_mapped_df = pd.read_table(EPD_TATA_mapped, sep='\t', header=None)
cols = ['chr','start','stop','promoter_AGI','dot1A','strand','sourceA','typeA','dot2A','attributes','chrB','startB','stopB','typeB','remove_col','strandB','bp_overlap']
EPD_TATA_mapped_df.columns = cols
```

```python
EPD_TATA_mapped_df = EPD_TATA_mapped_df[['chr','start','stop','promoter_AGI','dot1A','strand','sourceA','typeA','dot2A','attributes','chrB','startB','stopB','strandB']]
```

```python
#TATA boxes in both EPD and also from scanning with FIMO containing TBP2
merged_TATA = pd.merge(merged[merged.TF_AGI == 'AT1G55520'], EPD_TATA_mapped_df, on='promoter_AGI', how='left')
```

```python
#filter nan
merged_TATA = merged_TATA[merged_TATA.attributes.notnull()].reset_index(drop=True)
```
